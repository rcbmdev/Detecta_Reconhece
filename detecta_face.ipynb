```python
import cv2
from skimage import io, color
from IPython.display import Image
```

```python
Image(url="Imagens/faces.jpg")
```

```python
img_face1 = io.imread("Imagens/faces.jpg")
io.imshow(img_face1)
io.show()
```

```python
img_face1 = cv2.imread("Imagens/faces.jpg")
img_face1.shape
```

```python
img_face1_gray = color.rgb2gray(img_face1)
io.imshow(img_face1_gray)
io.show()
```

```python
img_face1_gray.shape
```

```python
grayImage1 = cv2.cvtColor(img_face1, cv2.COLOR_BGR2GRAY)
grayImage1.shape
```

```python
face_parametro = "Cascade/frontalface_alt.xml"
classificador_cascade = cv2.CascadeClassifier(face_parametro)
```

```python
classificador_cascade
```

```python
faces = classificador_cascade.detectMultiScale(grayImage1, scaleFactor=1.2, minNeighbors=10, minSize=(30, 30))
```

```python
faces
```

```python
for (x, y, width, height) in faces:
    cv2.rectangle(img_face1, (x, y), (x +width, y+height), (0, 255, 0), 6 )
```

```python
cv2.imwrite("Imagens/faces_resultado.jpg", img_face1)
```
