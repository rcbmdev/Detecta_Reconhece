```python
import face_recognition
import PIL.Image
import PIL.ImageDraw
```

```python
imagem = face_recognition.load_image_file("Imagens/faces.jpg")
```

```python
faces_encontradas = face_recognition.face_locations(imagem)
len(faces_encontradas)
```

```python
faces_encontradas
```

```python
image_array = PIL.Image.fromarray(imagem)
for face in faces_encontradas:
    top, left, bottom, right = face
    if left > right:
        left, right = right, left
    if top > bottom:
        top, bottom = bottom, top
    draw = PIL.ImageDraw.Draw(image_array)
    draw.rectangle([left, top, right, bottom], outline="green", width=15)
```
